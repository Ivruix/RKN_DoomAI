```python
from annoy import AnnoyIndex
import json

# Загружаем Annoy-индекс
embedding_dim = 512  # Укажите размерность, как в исходном скрипте
index = AnnoyIndex(embedding_dim, 'angular')
index.load("/kaggle/input/rkn-first-annoy/image_embeddings.ann")

# Загружаем метаданные
with open("/kaggle/input/rkn-first-annoy/annoy_metadata.json", 'r') as f:
    metadata = json.load(f)
```

```python
n_trees = 1024

new_index = AnnoyIndex(embedding_dim, 'angular')
n_items = index.get_n_items()
for i in range(n_items):
    new_index.add_item(i, index.get_item_vector(i))
new_index.build(n_trees)
index.save(f"image_embeddings_{n_trees}.ann")
```
